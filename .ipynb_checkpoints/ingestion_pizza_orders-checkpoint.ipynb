## Ingestão de arquivo CSV com pedidos da pizzaria no PostgreSQL    
<br/>

* **Tabela destino:** maven_pizza.raw_pizza_orders
* **Recorrência:** todos os dias às 3AM BRT.
* **Utilizado em:** dashboard de análise de vendas de pizza [(link aqui)](https://www.google.com).

```python
import pandas as pd
import psycopg2
import json
from sqlalchemy import create_engine

# carregar arquivo json com as credenciais do banco de dados
db_credentials_file = open('psql.json')
db_credentials      = json.load(db_credentials_file)

# ler os detalhes de conexão
db_host     = db_credentials['db_host']
db_port     = db_credentials['db_port']
db_name     = db_credentials['db_name']
db_user     = db_credentials['db_user']
db_password = db_credentials['db_password']

# criar a engine para rodar os comandos
db_engine = create_engine('postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'.format(
      db_host     = db_host     
    , db_port     = db_port     
    , db_name     = db_name     
    , db_user     = db_user     
    , db_password = db_password 
))
```

```python
# ler arquivo csv a ser inserido
raw_pizza_orders_path = "raw_pizza_orders.csv"
df_raw_pizza_orders = pd.read_csv(raw_pizza_orders_path, sep = ";")
df_raw_pizza_orders.iloc[0:5]
```

```python
# criar uma tabela no banco chamada "raw_pizza_orders" utilizando o arquivo csv
df_raw_pizza_orders.to_sql(name = 'raw_pizza_orders', con = db_engine, if_exists = 'replace', index = False)
```
